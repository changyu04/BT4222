```python
import pandas as pd
import numpy as np
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from sklearn.metrics import classification_report
from nltk.corpus import sentiwordnet as swn
```

# VADER for aspect based sentiment analysis
lexicon rule based model

https://github.com/cjhutto/vaderSentiment. 
- there is no need for validation using VADER, but for fair comparision, we will just run the model on our testing set (same test set as our other supervised models)

```python
test = pd.read_csv("data/test.csv").drop("Unnamed: 0", axis = 1)

# ground truth
food_sentiment = test["food_sentiment"]
service_sentiment = test["service_sentiment"]
ambience_sentiment = test["ambience_sentiment"]

#combining
test
```

```python
#VADER.make_lex_dict() The valence aware dictionary is quite limited here but cant seem to find free valence aware dictionary online to update this dictionary
```

```python

VADER = SentimentIntensityAnalyzer()

def extract_scores(text):
    sentiment_dict = VADER.polarity_scores(text)
    # we rank the scores, the predicted class would be the line with the highest score
    positive_score = sentiment_dict["pos"]
    neg_score = sentiment_dict["neg"]
    neu_score = sentiment_dict["neu"]

    # prioritize ties in the order positive > negative > neutral (bit biased but...)
    if positive_score >= neg_score and positive_score >= neu_score:
        return 1
    elif neg_score >= positive_score and neg_score >= neu_score:
        return -1
    elif neu_score >= positive_score and neu_score >= neg_score:
        return 0
```

```python
#print(len(VADER.make_lex_dict().keys()))

# update the valence aware dictionary

#VADER.polarity_scores(test.food_aspect_extraction_1[0])
results_table = {
    "model_name" : [],
    "accuracy" : [],
    "weighted_f1" :[],
    "NEG_precision" : [],
    "NEG_recall" : [],
    "NEU_precision" :[],
    "NEU_recall" : [],
    "POS_precision":[],
    "POS_recall" : []
}
```

# For aspect extraction 1

```python
#replace the NaN's with empty string
test.loc[test.food_aspect_extraction_1.isna() ,"food_aspect_extraction_1"] = "normal"
food_aspect_extraction_1 = test.food_aspect_extraction_1

test.loc[test.service_aspect_extraction_1.isna() ,"service_aspect_extraction_1"] = "normal"
service_aspect_extraction_1 = test.service_aspect_extraction_1

test.loc[test.ambience_aspect_extraction_1.isna() ,"ambience_aspect_extraction_1"] = "normal"
ambience_aspect_extraction_1 = test.ambience_aspect_extraction_1
```

```python
# food aspect predictions and classification results using aspect extraction 1
prediction  = food_aspect_extraction_1.apply(lambda x : extract_scores(x))    
res = classification_report(food_sentiment,prediction, output_dict= True)

results_table["model_name"].append("VADER_food_extraction_1")
results_table["accuracy"].append(res["accuracy"])
results_table["weighted_f1"].append(res["weighted avg"]["f1-score"])
results_table["NEG_precision"].append(res["-1"]["precision"])
results_table["NEG_recall"].append(res["-1"]["recall"])
results_table["NEU_precision"].append(res["0"]["precision"])
results_table["NEU_recall"].append(res["0"]["recall"])
results_table["POS_precision"].append(res["1"]["precision"])
results_table["POS_recall"].append(res["1"]["recall"])
```

```python
prediction  = service_aspect_extraction_1.apply(lambda x : extract_scores(x))    

res = classification_report(service_sentiment,prediction, output_dict= True)
results_table["model_name"].append("VADER_service_extraction_1")
results_table["accuracy"].append(res["accuracy"])
results_table["weighted_f1"].append(res["weighted avg"]["f1-score"])
results_table["NEG_precision"].append(res["-1"]["precision"])
results_table["NEG_recall"].append(res["-1"]["recall"])
results_table["NEU_precision"].append(res["0"]["precision"])
results_table["NEU_recall"].append(res["0"]["recall"])
results_table["POS_precision"].append(res["1"]["precision"])
results_table["POS_recall"].append(res["1"]["recall"])
```
